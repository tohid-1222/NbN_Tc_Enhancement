```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os
```

```python
pwd = os.getcwd()
print(pwd)
```

```python
df_method = pd.read_excel(pwd + '/box_plots/Boxplot_data_methods.xlsx')
df_method.head()
```

```python
df_substrate = pd.read_excel(pwd + '/box_plots/Boxplot_data_substrates.xlsx')
df_substrate.head()
```

```python
box_df2 = df_substrate
box_x2 = box_df2['Substrate']
box_y2 = box_df2['Predicted_Tc(K)']
```

```python
df_exp = pd.read_excel(pwd + '/data/NbN_new_feature_added.xlsx')
df_exp.head()
```

```python
box_df3 = df_exp

box_df3 = box_df3[box_df3["substrate"].str.contains("MgF2") == False]
box_df3 = box_df3[box_df3["substrate"].str.contains("HfO2") == False]

box_x3 = box_df3['substrate']
box_y3 = box_df3['Tc_K']
```

```python
box_x4 = box_df3['method']
box_y4 = box_df3['Tc_K']
```

```python
df = pd.read_excel(pwd + '/actual_prediction_df.xlsx')
df.head()
```

```python
df2 = df

df2 = df2[df2["substrate"].str.contains("MgF2") == False]
df2 = df2[df2["substrate"].str.contains("HfO2") == False]

df2.head()
```

```python
box_x5 = df2['method']
box_y5 = df2['predicted_Tc']
```

```python
box_x6 = df2['substrate']
box_y6 = df2['predicted_Tc']
```

```python
fig, axes = plt.subplots(1, 2, sharex=False, figsize=(12,6))

axes[0].set_title('Experimental $T_{C}$ vs substrates')
axes[1].set_title('Predicted $T_{C}$ vs substrates')

axes[0].set_ylim(7.0, 17.0)
axes[1].set_ylim(7.0, 17.0)

sns.boxplot(ax=axes[0], x = box_x3,
            y = box_y3,
            saturation=2.0,
            data = box_df3)
sns.stripplot(ax=axes[0], x = box_x3,
              y = box_y3,
              color = 'black',
              data = box_df3)


sns.boxplot(ax=axes[1], x = box_x6,
            y = box_y6,
            saturation=2.0,
            data = df2)
sns.stripplot(ax=axes[1], x = box_x6,
              y = box_y6,
              color = 'black',
              data = df2)


axes[0].set_xlabel('Substrate')
axes[1].set_xlabel('Substrate')

axes[0].set_ylabel('Experimental $T_{c}$ (K)')
axes[1].set_ylabel('Predicted $T_{c}$ (K)')

plt.tight_layout()
#plt.savefig(pwd+'/figures/Exp_Pred_Tc_vs_Substrates.png', dpi=500)
plt.show()
```
