```python

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
```

```python
pwd = os.getcwd()
print(pwd)
```

```python
sbstrts = ['$Al_2O_3$', 'GaAs', '$LiNbO_3$', '$MgF_2$', 'MgO', 'Si', 'SiN', '$HfO_2$']
exp = [16.00, 14.60, 15.20, 9.00, 16.10, 15.07, 12.00, 13.10]
predicted = [16.05, 15.30, 16.47, 16.58, 16.73, 16.14, 10.12, 15.30]

exp_temp = [850, 400, 800, 25, 400, 800, 600, 300]
exp_pressure = [1.0, 0.33, 0.08, 0.00002, 0.33, 66.67, 9.33, 53.3]
exp_thick = [27, 140, 5, 5, 140, 25, 200, 264]
```

```python
df = pd.DataFrame(
{
    'Max Tc(exp) (K)': exp,
    'Temperature (C)': exp_temp,
    'Pressure (Pa)': exp_pressure,
    'Thickness (nm)': exp_thick,
    'Max Tc(predicted) (K)': predicted
}
)
print(df.shape)
#df.to_excel(pwd+'/comparison_variable_data.xlsx', index=False)
df.head()
```

```python
# set width of bar
barWidth = 0.25
fig = plt.subplots(figsize =(16, 8))

# Set position of bar on X axis
br1 = np.arange(len(sbstrts))
br2 = [x + barWidth for x in br1]

# Make the plot
bar1 = plt.bar(br1, exp, color ='r', width = barWidth,
        edgecolor ='grey', label ='Max Tc in observed data')
bar2 = plt.bar(br2, predicted, color ='g', width = barWidth,
        edgecolor ='grey', label ='Predicted max Tc')


plt.ylim(0, 26)
# Adding Xticks
plt.xlabel('Substrate', fontsize = 15)
plt.ylabel('$T_{C}$ (K)', fontsize = 15)
plt.xticks([r + barWidth/2 for r in range(len(sbstrts))], sbstrts)
plt.legend()

for bar in bar1:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .2, yval, fontsize=10)
for bar in bar2:
    yval = bar.get_height()
    plt.text(bar.get_x(), yval + .2, yval, fontsize=10)
plt.savefig(pwd + '/Comparison_plot.png', dpi=650, bbox_inches="tight")
plt.show()
```
